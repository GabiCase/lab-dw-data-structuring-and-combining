# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
url='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
pd.set_option('display.max_columns', None)

df=pd.read_csv(url)
df
```

```python
#eliminar columnas que no aportan nada

df.drop(columns=['unnamed:_0'], inplace=True)
df
```

```python
#comprobar que no haya filas duplicadas
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True, ignore_index=True)
df['customer'].duplicated().sum()


duplicados = df[df['customer'].duplicated()]['customer'].value_counts()
duplicados

# Definir la función que obtiene el valor más frecuente
def most_frequent(series):
    return series.mode().iloc[0] if not series.mode().empty else pd.NA

# Agrupar por 'client_id' y aplicar la función de agregación
result = df.groupby('customer').agg(most_frequent)

# Restablecer el índice si quieres que 'client_id' vuelva a ser una columna
df = result.reset_index()

df
```

```python
df['customer'].duplicated().sum()
df.columns
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_table_sales = df.pivot_table(
    values='total_claim_amount',
    index='sales_channel',
    aggfunc='sum'
)

# Redondear a 2 decimales
pivot_table_sales = pivot_table_sales.round(2)

# Ordenar de mayor a menor
pivot_table_sales = pivot_table_sales.sort_values(by='total_claim_amount', ascending=False)

# Mostrar la tabla pivot
print(pivot_table_sales)
```

```python
pivot_table_value = df.pivot_table(
    values='customer_lifetime_value',
    index=['gender','education'],
    aggfunc='sum'
)

# Redondear a 2 decimales
pivot_table_value = pivot_table_value.round(2)

# Ordenar de mayor a menor
pivot_table_value = pivot_table_value.sort_values(by='customer_lifetime_value', ascending=False)

# Mostrar la tabla pivot
print(pivot_table_value)
```
